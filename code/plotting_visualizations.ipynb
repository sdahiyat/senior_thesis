```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSVs
nuc_df = pd.read_csv("/mnt/lareaulab/sdahiyat/Corrected_Nucleotide_Shuffling_Results.csv")
benchmark_df = pd.read_csv("/mnt/lareaulab/sdahiyat/spliceai_benchmark_results.csv")

# Extract AUPRC values
benchmark_auprc = benchmark_df[benchmark_df["Metric"] == "AUPRC"]
nuc_auprc = nuc_df[nuc_df["Metric"] == "AUPRC"]

# Calculate AUPRC loss: benchmark - shuffled
auprc_loss = [
    benchmark_auprc["Acceptor_400"].values[0] - nuc_auprc["Acceptor_400"].values[0],
    benchmark_auprc["Acceptor_2k"].values[0] - nuc_auprc["Acceptor_2k"].values[0],
    benchmark_auprc["Donor_400"].values[0] - nuc_auprc["Donor_400"].values[0],
    benchmark_auprc["Donor_2k"].values[0] - nuc_auprc["Donor_2k"].values[0],
]

# Group labels and colors
labels = ["Acceptor 400", "Acceptor 2000", "Donor 400", "Donor 2000"]
x = np.arange(len(labels))
colors = ["#FDB515", "#003262", "#FDB515", "#003262"]  # gold for 400, blue for 2000

# Plot AUPRC loss
plt.figure(figsize=(10, 6))
bars = plt.bar(x, auprc_loss, color=colors, alpha=0.95)
plt.xticks(x, labels, rotation=45)
plt.ylabel("AUPRC Loss")
plt.title("Loss of AUPRC After Shuffling Nucleotides")
plt.grid(axis="y", linestyle="--", alpha=0.6)

# Annotate values
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.01, f"{height:.3f}", ha='center', va='bottom')

# Save both PNG and SVG
plt.tight_layout()
plt.savefig("/mnt/lareaulab/sdahiyat/illumina/poster_figures/auprc_loss_nucleotide_shuffling.png", dpi=300)
plt.savefig("/mnt/lareaulab/sdahiyat/illumina/poster_figures/auprc_loss_nucleotide_shuffling.svg", format='svg')
plt.show()
```
